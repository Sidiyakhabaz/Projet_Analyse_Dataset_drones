```python
import pandas as pd
import numpy as np
import os


# Rechargement des données finales
final_path  = r"C:\Drone_Attack_Similarity_Project\DATASET\Final"
tables_path = r"C:\Drone_Attack_Similarity_Project\Rapport\tables"
os.makedirs(tables_path, exist_ok=True)

datasets_final = {}
for filename in os.listdir(final_path):
    if filename.endswith("_final.csv"):
        name = filename.replace("_final.csv", "").replace("_", "-")
        datasets_final[name] = pd.read_csv(f"{final_path}\\{filename}")

print("Datasets finaux chargés :")
for name, df in datasets_final.items():
    print(f"  {name:<18} → {df.shape[0]} lignes, {df.shape[1]} colonnes")

######### Mapping MCP — UAVIDS-2025 ########

# Mapping manuel basé sur la sémantique technique des features
MCP_MAPPING_UAVIDS = {

    # Mécanisme (M) — comment l'attaque s'exécute
    # → features décrivant le comportement du trafic réseau
    "M": [
        "TxPackets",
        "RxPackets",
        "LostPackets",
        "TxBytes",
        "RxBytes",
        "TxPacketRate/s",
        "RxPacketRate/s",
        "TxByteRate/s",
        "RxByteRate/s",
        "PacketDropRate",
    ],

    # Composant (C) — quoi est ciblé
    # → features identifiant les protocoles et adresses
    "C": [
        "Protocol",
        "SrcPort",
        "DstPort",
    ],

    # Paramètre (P) — avec quels paramètres temporels/réseau
    # → features décrivant la durée, le délai, le débit
    "P": [
        "FlowDuration/s",
        "MeanDelay/s",
        "MeanJitter/s",
        "Throughput/Kbps",
        "MeanPacketSize",
        "AverageHopCount",
    ],
}

######### Mapping MCP — Composite IoT/UAV (9 datasets) ########

# Les 9 datasets individuels ont les mêmes colonnes
# donc un seul mapping s'applique à tous
MCP_MAPPING_COMPOSITE = {

    # Mécanisme (M)
    "M": [
        "Total Fwd Packets",
        "Total Backward Packets",
        "Total Length of Fwd Packets",
        "Total Length of Bwd Packets",
        "Fwd Packet Length Max",
        "Fwd Packet Length Min",
        "Fwd Packet Length Mean",
        "Bwd Packet Length Max",
        "Bwd Packet Length Min",
        "Bwd Packet Length Mean",
        "Flow Packets/s",
        "Fwd Packets/s",
        "Bwd Packets/s",
    ],

    # Composant (C)
    "C": [
        "Protocol",
        "Source Port",
        "Destination Port",
        "Fwd Header Length",
        "Bwd Header Length",
    ],

    # Paramètre (P)
    "P": [
        "Flow Duration",
        "Flow IAT Mean",
        "Flow IAT Std",
        "Flow IAT Max",
        "Flow IAT Min",
        "Fwd IAT Total",
        "Fwd IAT Mean",
        "Bwd IAT Total",
        "Bwd IAT Mean",
        "Flow Bytes/s",
        "Average Packet Size",
        "Avg Fwd Segment Size",
        "Avg Bwd Segment Size",
    ],
}

######### Vérification du mapping ########

print("\n Mapping UAVIDS-2025")
for dim, features in MCP_MAPPING_UAVIDS.items():
    print(f"\n  Dimension {dim} ({len(features)} features) :")
    for f in features:
        print(f"    - {f}")

total_uavids = sum(len(v) for v in MCP_MAPPING_UAVIDS.values())
print(f"\n  Total features mappées : {total_uavids}")

print("\n Mapping Composite IoT/UAV ")
for dim, features in MCP_MAPPING_COMPOSITE.items():
    print(f"\n  Dimension {dim} ({len(features)} features) :")
    for f in features:
        print(f"    - {f}")

total_composite = sum(len(v) for v in MCP_MAPPING_COMPOSITE.values())
print(f"\n  Total features mappées : {total_composite}")

######### Export du mapping en CSV ########

# UAVIDS-2025
rows_uavids = []
for dim, features in MCP_MAPPING_UAVIDS.items():
    for f in features:
        rows_uavids.append({"Feature": f, "Dimension": dim, "Dataset": "UAVIDS-2025"})

df_mapping_uavids = pd.DataFrame(rows_uavids)
df_mapping_uavids.to_csv(f"{tables_path}\\mapping_MCP_UAVIDS.csv", index=False)

# Composite IoT/UAV
rows_composite = []
for dim, features in MCP_MAPPING_COMPOSITE.items():
    for f in features:
        rows_composite.append({"Feature": f, "Dimension": dim, "Dataset": "Composite IoT/UAV"})

df_mapping_composite = pd.DataFrame(rows_composite)
df_mapping_composite.to_csv(f"{tables_path}\\mapping_MCP_Composite.csv", index=False)

# Mapping complet combiné
df_mapping_all = pd.concat([df_mapping_uavids, df_mapping_composite], ignore_index=True)
df_mapping_all.to_csv(f"{tables_path}\\mapping_MCP_complet.csv", index=False)

print("\n Mapping exporté dans Rapport/tables/")
print(f"   → mapping_MCP_UAVIDS.csv    ({len(df_mapping_uavids)} features)")
print(f"   → mapping_MCP_Composite.csv ({len(df_mapping_composite)} features)")
print(f"   → mapping_MCP_complet.csv   ({len(df_mapping_all)} features au total)")
```
